```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'LabAssignment_Day3/Dataset_Day3.csv'
data = pd.read_csv(file_path)

# Calculate the distance

data['distance'] = np.sqrt((data['dropoff_latitude'] - data['pickup_latitude'])**2 + (data['dropoff_longitude'] - data['pickup_longitude'])**2)

# Detecting outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Find outliers for fare_amount, passenger_count, and distance
outliers_fare = detect_outliers(data, 'fare_amount')
outliers_passenger = detect_outliers(data, 'passenger_count')
outliers_distance = detect_outliers(data, 'distance')

# Combine all outliers
all_outliers = pd.concat([outliers_fare, outliers_passenger, outliers_distance]).drop_duplicates()

# Remove outliers from the dataset
cleaned_data = data[~data['key'].isin(all_outliers['key'])]

# Scatterplot between distance and fare_amount
plt.figure(figsize=(10, 6))
plt.scatter(cleaned_data['distance'], cleaned_data['fare_amount'], alpha=0.5)
plt.title('Scatterplot between Distance and Fare Amount')
plt.xlabel('Distance')
plt.ylabel('Fare Amount')
plt.grid(True)
plt.show()
```
